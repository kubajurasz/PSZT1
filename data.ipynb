```python
import matplotlib.pyplot as plt
```

```python
def bar_plot(x_val, y_val, name, xlabel, ylabel):
    plt.figure(figsize=(12,8))
    plt.bar(x_val, y_val)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.savefig(name)
    plt.show()
```

```python
def load_data(file):
    dupsko = []
    with open(file) as file:
        for line in file:
            temp = line.strip().split(",")
            if temp != ['']:
                dupsko.append([float(i) for i in temp])
    return dupsko
```
